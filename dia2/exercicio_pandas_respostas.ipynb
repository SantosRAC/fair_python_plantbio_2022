## 16. Crie uma coluna com a área da pétala

OBS: considere que a pétala é um retângulo

```python
iris_df['petal_area'] = iris_df['petallength'] * iris_df['petalwidth']
iris_df
```

## 17. Filtre o dataframe para áreas de pétala com valor entre 7 e 9, e ordene os valores de maior para menor

```python
# primeiro filtro (valor > 7)
filtro_maior_7 = iris_df[iris_df['petal_area'] > 7]
filtro_maior_7
```

```python
# segundo filtro (valor < 9)
filtro_menor_9 = filtro_maior_7[filtro_maior_7['petal_area'] < 9]
filtro_menor_9
```

```python
# ordenar do maior para o menor
# o df filtro_menor_9 contém os dois filtros
df_final = filtro_menor_9.sort_values('petal_area', ascending=False)
df_final
```

```python
# mesma resposta acima, mas utilizando apenas um comando (funções encadeadas)
iris_df[(iris_df['petal_area'] > 7) & (iris_df['petal_area'] < 9)].sort_values('petal_area', ascending=False)
```
